# **Project Description**

This project implements a Course Management System using object-oriented programming (OOP) principles in Python. It involves designing a system where various types of courses can be created, with specific attributes and behaviors, while leveraging key OOP concepts such as inheritance, encapsulation, polymorphism, and abstraction.

# **Objectives**

**Centralized Course Management:** Provide a platform for managing all course-related data.<br>
**Automated Enrollment:** Streamline student enrollment and course registration.<br>
**Syllabus Management:** Simplify the creation and distribution of course syllabi.<br>
**Scheduling:** Allow for structured management of lectures and lab sessions.<br>
**Custom Course Types:** Support multiple course types with specific attributes and schedules (e.g., Programming Fundamentals, DBMS).


### **Scenario Overview**
Imagine you're a student at COMSATS University, managing your various courses throughout a semester. This project contains that managing system. Each course, like **Programming Fundamentals**, **OOPS**, **Python**, **Statistics**, and **Calculus**, has its own specific characteristics and requirements. However, they also share common features.

### **Class: Course**
A **Course** as a general template that holds information about all courses at COMSATS.

### **Objects: Specific Courses**
Now, each specific course you take is an **object** of the **Course** class. For instance:
- **Programming Fundamentals** (object)
- **OOPS** (object)
- **Python** (object)

Each of these courses will have unique attributes and methods based on what they offer.

#### **Example of Attributes:**
- **Course Name**: The name of the course (e.g., "Programming Fundamentals").
- **Credits**: The number of credits the course is worth.
- **Instructor**: The name of the instructor teaching the course.


#### **Example of Methods:**
- **Get Syllabus**: A method that provides the course syllabus.
- **Enroll Student**: A method to enroll students in the course.



In this project, we will be covering different concepts of **Object Oriented Proframming** that will be helping us in managing the system in a very effective way.

### **1. Inheritance: Specialized Courses**
Inheritance allows courses to share common features while also having specific characteristics.

1. **Base Class: Course**
   - Common attributes: `course_name`, `credits`, `instructor`
   - Common methods: `get_syllabus()`, `enroll_student()`

2. **Derived Classes:**
   - **ProgrammingCourse** (inherits from Course):
     - Additional attribute: `programming_language`
     - Method: `get_practical_exercises()`
   
   - **MathematicsCourse** (inherits from Course):
     - Additional attribute: `math_field` (e.g., Calculus, Statistics)
     - Method: `get_formula_sheet()`


### **2. Polymorphism: Different Behaviors**
Polymorphism allows different types of courses to have a method that behaves differently based on the course type.

- Both **ProgrammingCourse** and **MathematicsCourse** can have a method called `get_schedule()`. However, the output will be different:
  - **ProgrammingCourse**: Might include coding assignments and lab sessions.
  - **MathematicsCourse**: Might include theory lectures and problem-solving sessions.



### **3. Encapsulation:**
- **Private Attributes**: These are the details of a class (like course information) that you don’t want everyone to change directly. You hide them by using a double underscore (`__`).
  - **Example**: `__course_name`, `__credits`, and `__instructor` in our **Course** class.

- **Public Methods**: These are the functions that allow you to access or change the private attributes safely. Think of them as the "safe doors" to get in and out of the private data.
  - **Example**: Methods like `get_course_name()`, `get_credits()`, and `enroll_student()` let you interact with the course details without exposing them directly.



### **4.Abstraction: Hiding Complexity**
Abstraction allows you to interact with a course without needing to understand all the underlying details.

- You can enroll in a course by simply calling the `enroll_student()` method without needing to know how the enrollment process works behind the scenes (like database operations, checks, etc.).

# **Super Class: Course**

```python
# Base Class Course()
class Course():
  def __init__(self, course_name, course_code, semester, department, credits, instructor, syllabus, lab):
    self.__course_name = course_name
    self.__course_code = course_code
    self.__semester = semester
    self.__department = department
    self.__credits = credits
    self.__lab = lab
    self.__instructor = instructor
    self.__syllabus = syllabus
    self.__students = {}
    self.__rolls = []
    self.__courses_enrollments = {}


  def get_course_name(self):
    return f'This is "{self.__course_name}" course.'

  def get_course_code(self):
    return f'The course code is: {self.__course_code}'

  def get_semester(self):
    return f'{self.__course_name} is being offered to {self.__semester} semester(s).'

  def get_department(self):
    return f'{self.__course_name} is offered in {self.__department} department(s).'

  def lab_exits(self):
    if self.__lab == True:
      return f'{self.__course_name} has lab too.'
    else:
      return f'{self.__course_name} has no lab.'

  def get_credits(self):
    return f'{self.__course_name} comprises of {self.__credits} credit hours.'

  def get_instructor(self):
    return f'{self.__instructor} is the course instructor.'

  def get_students(self):
    return f'The students enrolled in {self.__course_name} are: \n{self.__students}'

  def total_hrs_per_week(self):
    if self.__lab == 1:
      return f'{self.__course_name} has 3hrs theory and 3hrs lab per week.'
    else:
      return f'{self.__course_name} has 3hrs theory class per week.'

  # Abstract method
  def get_schedule(self):
    return

  # Common methods: get_syllabus(), enroll_student()
  def get_syllabus(self):
    return f'The syllabus comprises of: \n{self.__syllabus}'

  def enroll_student(self, student_name, roll_no):
    if roll_no not in self.__rolls:
      self.__students[roll_no] = student_name
      self.__rolls.append(roll_no)

    if self.__course_name not in self.__courses_enrollments:
      self.__courses_enrollments[self.__course_name] = {}

    if roll_no not in self.__courses_enrollments[self.__course_name]:
      self.__courses_enrollments[self.__course_name][roll_no] = student_name
      return f"{student_name} is enrolled in {self.__course_name}"
    else:
      return f"{student_name} is already enrolled in {self.__course_name}"
```

# **Derived Class: ProgrammingCourse**

```python
# Derived Class: ProgrammingCourse (inherits from Course):
class ProgrammingCourse(Course):
    def __init__(self, course_name, course_code, semester, department, credits, instructor, syllabus, programming_language, lab):
        super().__init__(course_name, course_code, semester, department, credits, instructor, syllabus, lab)
        # Additional attribute: programming_language
        self.__programming_language = programming_language

    def get_programming_language(self):
        return f"This course is in {self.__programming_language}."

    # Additional Method: get_practical_exercises()
    def get_practical_exercises(self):
        return f"Practical exercises are for {self.__programming_language}."

    # Polymorphism (Different behaviour for Programming Course)
    # Abstraction: overrite the get_schedule() method of super class Course
    def get_schedule(self):
        return "Includes coding assignments and lab sessions."
```

# **Derived Class: MathematicsCourse**

```python
# Derived Class: MathematicsCourse (inherits from Course):
class MathematicsCourse(Course):
    def __init__(self, course_name, course_code, semester, department, credits, instructor, syllabus, math_field, software_tool, lab = False):
        super().__init__(course_name, course_code, semester, department, credits, instructor, syllabus, lab)
        # Additional attribute: math_field (e.g., Calculus, Statistics)
        self.__math_field = math_field
        self.__software_tool = software_tool # e.g; MATLAB

    def get_math_field(self):
        return f"This course is related to {self.__math_field}."

    def get_software_tool(self):
        return f"The software tool used in this course is {self.__software_tool}."

    # Additional Method: get_formula_sheet()
    def get_formula_sheet(self):
        return f"Formula sheet for {self.__math_field} using {self.__software_tool}."

    # Polymorphism (Different behaviour for Mathematical Course)
    # Abstraction: overrite the get_schedule() method of super class Course
    def get_schedule(self):
        return "Includes theory lectures and problem-solving sessions."
```

# **Derived Class: PhysicsCourse**

```python
# Derived Class: PhysicsCourse (inherits from Course):
class PhysicsCourse(Course):
    def __init__(self, course_name, course_code, semester, department, credits, instructor, syllabus, physics_branch, lab):
        super().__init__(course_name, course_code, semester, department, credits, instructor, syllabus, lab)
        # Additional attribute: physics_branch
        self.__physics_branch = physics_branch

    def get_physics_branch(self):
        return f"This course is related to {self.__physics_branch}."

    # Polymorphism (Different behaviour for Physics Course)
    # Abstraction: overrite the get_schedule() method of super class Course
    def get_schedule(self):
        return "Includes theory lectures and practical experimentation sessions."
```

# **Derived Class: HumanitiesCourse**

```python
# Derived Class: HumanitiesCourse (inherits from Course):
class HumanitiesCourse(Course):
    def __init__(self, course_name, course_code, semester, department, credits, instructor, syllabus, humanities_field, lab=False):
        super().__init__(course_name, course_code, semester, department, credits, instructor, syllabus, lab)
        # Additional attribute: Humanities Field
        self.__humanities_field = humanities_field

    def get_humanities_field(self):
        return f"This course is related to {self.__humanities_field}."

    # Polymorphism (Different behaviour for Humanities Course)
    # Abstraction: overrite the get_schedule() method of super class Course
    def get_schedule(self):
        return "Includes theory lectures and hands-on practice sessions."
```

# **Displaying Courses' Information**
